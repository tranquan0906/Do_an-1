```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['sellingprice'], kde=True)
plt.title('Phân phối giá bán xe')
plt.xlabel('Giá bán (USD)')
plt.ylabel('Tần suất')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='year', y='sellingprice', data=df)
plt.title('Giá bán theo năm sản xuất')
plt.xlabel('Năm sản xuất')
plt.ylabel('Giá bán (USD)')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['odometer'], df['sellingprice'], alpha=0.3)
plt.title('Mối quan hệ giữa số km đã đi và giá bán')
plt.xlabel('Số km đã đi')
plt.ylabel('Giá bán (USD)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['make'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 hãng xe phổ biến nhất')
plt.xlabel('Hãng xe')
plt.ylabel('Số lượng')
plt.show()
```

```python
top_makes = df['make'].value_counts().head(10).index
plt.figure(figsize=(10, 6))
df[df['make'].isin(top_makes)].groupby('make')['sellingprice'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Giá trung bình theo hãng xe (Top 10)')
plt.xlabel('Hãng xe')
plt.ylabel('Giá trung bình (USD)')
plt.show()
```

```python
numeric_cols = ['year', 'odometer', 'mmr', 'sellingprice']
plt.figure(figsize=(8, 6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Ma trận tương quan')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['seller'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Đơn vị bán được nhiều nhất')
plt.xlabel('Đơn vị bán')
plt.ylabel('Số lượng xe bán ra')
plt.show()
```
